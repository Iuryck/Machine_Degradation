## Pegando pedacos de DataFrames
###### Agora queremos criar um dataframe unicamente com os dados relativos a regiao de Albany, entao o que fazemos e criar um dataframe com nome diferente (albany_df) e associa aos dados. 


```python
albany_df = df[ df['region'] == 'Albany']

#    Mostramos que agora todas as fileiras vao ter Albany na coluna Region
albany_df.head()
```

##### Este dataframe tem um problema em que o indice sao apenas numeros sem sentido, se queremos relacionar direito nossos dados queremos uma coluna que as relacione bem como Datas. Para ver o indice fazemos
## Mexendo com o indice

```python
albany_df.index
```

```python
#   Para o pandas reconhecer as datas em formato de datas
albany_df['Date'] = pd.to_datetime(df['Date'])

#   Agora para substituir o indice pela coluna de Datas fazemos
albany_df.set_index('Date', inplace=True)
```

```python
#   Agora se queremos ver um grafico em funcao do indice/tempo fazemos
albany_df.plot()
```

##  Mostrando o grafico de uma coluna especifica

```python
albany_df['AveragePrice'].plot()
```

###### Os dados parecem volumosos, entao o que queremos fazer para suavizar a curva e uma "media movel", onde para cada ponto da curva uma quantia de dados anteriores  serao usados para fazer uma media. Fazendo assim uma media ao longo da curva toda
## Media Movel

```python
          
albany_df['AveragePrice'].rolling(25).mean().plot()
#.rolling define a janela de pontos de dados para cada ponto
#.mean faz a media de todos esses dados
```

    Como podemos ver algo deu errado. Parece que os dados tem algum problema. Vejamos

```python
albany_df.index
```

    O os dados comecam em 2015, vao retrocedendo, e depois volta para 2018 no final. Entao nosso dados estao fora de ordem
## Para botar os dados em ordem


```python
                    #Substituir o indice atual = verdadeiro
albany_df.sort_index(inplace=True)
albany_df['AveragePrice'].rolling(25).mean().plot()
```

    Esses dados podem ser uteis ao ponto de criar uma coluna so para ela, para que possamos analisar eles futuramente
## Criando Colunas

```python
albany_df['25ma'] = albany_df['AveragePrice'].rolling(25).mean()
albany_df['25ma'].plot()
```

    Observe que perdemos dados no inicio do grafico. Isso e pelo fato de nao existir a janela de 25 dados nesse inicio, entao nao tem como fazermos a media. Podemos ingorar essa exclusao de dados deixando explicito que nao deve haver um minimo de dados para fazer a media e entao os dados retornarao. Porem esses primeiros 25 pontos do grafico nao terao media e estarao desproporcionais com o restante do grafico.

```python
albany_df['25ma'] = albany_df['AveragePrice'].rolling(25, min_periods=0).mean()
albany_df['25ma'].plot()
```

    Os avisos estao dizendo que estamos alterando dados de um pedaco que retiramos de um dataframe maior, e que isso pode complicar nossa analise no futuro. Para evitar esses problemas e avisos podemos copiar um data frame
## Copiando DataFrames

```python
albany_df = df.copy()[df['region'] == 'Albany']

#Repitindo a construcao desse dataframe ate agora
albany_df.set_index('Date', inplace=True)
albany_df.sort_index(inplace=True)
albany_df['25ma'] = albany_df['AveragePrice'].rolling(25).mean()
albany_df['25ma'].plot()
```

## Mexendo com colunas
    Podemos pegar os valores de apenas uma coluna como um objeto...

```python
df['region'].values
```

    ...E depois transformamos em lista.

```python
df['region'].values.tolist()
```

    Pegamos muitos dados mas so queremos vizualizar os valores unicos, ou seja, pegar apenas quais valores estao na coluna, excluindo copias. Fazemos isso com o metodo Unique

```python
df['region'].unique()
```

    Agora digamos que 

```python
graph_df = pd.DataFrame()

#  Iterar por cada regiao que temos no dataframe
for region in df['region'].unique():
    #criando um dataframe de dados para cada regiao
    region_df = df.copy()[df['region']==region]
    #criando a coluna de media movel para cada dataframe de cada regiao
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    
    #Para as colunas nao se sobreporem damos o nome da regiao em cada coluna para 
    #que tenhamos uma coluna unica a cada regiao --> (f'{region}_)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    #Se o graph dataframe tiver vazio ele recebera o atual region_df
    if graph_df.empty:
        #para retornar um dataframe e nao uma serie usamos as chaves duplas
        graph_df = region_df[['price25ma']]
    
    else:
        graph_df = graph_df.join(region_df['price25ma'])
```
